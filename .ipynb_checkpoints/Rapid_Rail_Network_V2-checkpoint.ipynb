## Introduction ##

For the purpose of network analysis under the Network Science course, the dataset of **Rapid Rail Explorer** will be used. This dataset is owned by Prasarana Malaysia. Prasarana is the owner-operator of the country's rail services, including LRT networks, KL monorail and the MRT lines. In addition to that, they are also the owner-operator for the stage bus services in Kuala Lumpur, Selangor, Penang and Pahang. All these comprise what is commonly known as Rapid Rail and Rapid Bus systems. This dataset provides granular ridership data between Rapid Rail stations, including LRT networks, KL monorail and MRT lines. Analyzing passenger flows between stations can reveal patterns in public transportation usage.

This dataset can be acquired through url https://data.gov.my/dashboard/rapid-explorer. Though, the dataset must be retrieve programmatically via the API.

## Data Exploratory ##

The Rapid Rail Explorer data is accessible through the **GTFS Statis API** which provides standardized public transportation schedules and geographic information. The following is the code to retrieve the data using API:

```python
import requests
import pandas as pd
import zipfile
import io
import os

# define the API endpoint
api_url = "https://api.data.gov.my/gtfs-static/prasarana?category=rapid-rail-kl"

# request to the API
response = requests.get(api_url)

# check if the request was successful
if response.status_code == 200:
    # load the content into BytesIO object
    z = zipfile.ZipFile(io.BytesIO(response.content))
    # extract all files into current directory
    z.extractall()
    print("Files extracted successfully.")
    # load specific GTFS files into pandas DataFrames
    stops_df = pd.read_csv('stops.txt')
    trips_df = pd.read_csv('trips.txt')
    stop_times_df = pd.read_csv('stop_times.txt')
    routes_df = pd.read_csv('routes.txt')
    frequencies_df = pd.read_csv('frequencies.txt')
    # save each DataFrame to a csv file
    stops_df.to_csv("stops.csv", index = False)
    trips_df.to_csv("trips.csv", index = False)
    stop_times_df.to_csv("stop_times.csv", index = False)
    routes_df.to_csv("routes.csv", index = False)
    frequencies_df.to_csv("frequencies.csv", index = False)
    print("Data successfully saved to CSV files.")
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")    
```

There will be a set of files extracted from the API. Then all those dataset need to be combined as follows:

```python
# load the extracted csv files
stops_df = pd.read_csv('stops.csv')
trips_df = pd.read_csv('trips.csv')
stop_times_df = pd.read_csv('stop_times.csv')
routes_df = pd.read_csv('routes.csv')
frequencies_df = pd.read_csv('frequencies.csv')

# merge based on common columns
combined_df = trips_df.merge(routes_df, on = 'route_id', how = 'left')
combined_df = combined_df.merge(stop_times_df, on = 'trip_id', how = 'left')
combined_df = combined_df.merge(stops_df, on = 'stop_id', how = 'left')
combined_df = combined_df.merge(frequencies_df, on = 'trip_id', how = 'left')

# save the combined DataFrame to a csv file
combined_df.to_csv('combined_gtfs_data.csv', index = False)

print("All csv files successfully combined into 'combined_gtfs_data.csv'")
```

```python
data = pd.read_csv("combined_gtfs_data.csv")
data.head()
```

### 1. Selecting a relevant columns for network analysis ###

After combining all the dataset, there are a **total of 33 columns**. The prepare the dataset for network analysis, the data should be cleaned and transformed to contruct a meaningful relationship between stations to identify the vertices, edges and weight for the network system. The selected columns will then be saved as new dataset called **rapid_rail_network_data.csv**. The code to construct the relevant columns are as follows:

```python
# extracting unique stations for vertices
vertex_df = data[['stop_id', 'stop_name', 'stop_lat', 'stop_lon', 'isOKU']].drop_duplicates().reset_index(drop = True)

# prepare the edges
data_sorted = data.sort_values(by =['trip_id', 'stop_sequence'])

# create edges
data_sorted['to_stop_id'] = data_sorted['stop_id'].shift(-1)
data_sorted['to_stop_sequence'] = data_sorted['stop_sequence'].shift(-1)
data_sorted['next_trip_id'] = data_sorted['trip_id'].shift(-1)

# filter to ensure edges are within same trip and sequential
edges_df = data_sorted[
    (data_sorted['trip_id'] == data_sorted['next_trip_id']) &
    (data_sorted['stop_sequence'] + 1 == data_sorted['to_stop_sequence'])
][['trip_id', 'route_id', 'stop_id', 'to_stop_id', 'headway_secs']]

edges_df.rename(columns ={'stop_id': 'from_stop_id', 'to_stop_id': 'to_stop_id'}, inplace = True)
```

```python
vertex_df.head()
```

```python
edges_df.head()
```

```python
# combined the vertex and edges tables
edges_with_vertex_data = edges_df.merge(vertex_df, left_on = 'from_stop_id', right_on = 'stop_id', suffixes =('', '_from'))
edges_with_vertex_data = edges_with_vertex_data.merge(vertex_df, left_on = 'to_stop_id', right_on = 'stop_id', suffixes = ('_from', '_to'))

comprehensive_network_df = edges_with_vertex_data[
    ['trip_id', 'route_id', 'from_stop_id', 'stop_name_from', 'stop_lat_from', 'stop_lon_from', 'isOKU_from',
    'to_stop_id', 'stop_name_to', 'stop_lat_to', 'stop_lon_to', 'isOKU_to', 'headway_secs']
]
```

```python
comprehensive_network_df.head()
```

### 2. Determining the Vertices and Edges of the network ###

### Vertices : Name of the Station (from_station and to-station) ###

### Edges : Direct connection between stations ###

### Weight : travel frequency between stations (headway_secs) ###
The table are as below:

```python
rail_network_df = comprehensive_network_df[
    ['stop_name_from', 'stop_name_to', 'headway_secs']
].rename(columns = {
    'stop_name_from': 'from_station',
    'stop_name_to': 'to_station',
    'headway_secs': 'weight'
}).drop_duplicates().reset_index(drop = True)

rail_network_df.head()
```

```python
# save to CSV file
rail_network_df.to_csv('rail_network_data.csv', index = False)
```

### Visualize the Rail Network ###

```python
import networkx as nx
import matplotlib.pyplot as plt

# create edges 

# initialize the graph
G = nx.DiGraph()

# add edges
for _, row in rail_network_df.iterrows():
    G.add_edge(row['from_station'], row['to_station'], weight = row['weight'])

# number of vertices and edges
num_vertices = G.number_of_nodes()
num_edges = G.number_of_edges()

# visualize the graph
plt.figure(figsize = (14, 14))
pos = nx.spring_layout(G, seed = 42)

node_sizes = [300 + 1000 * nx.degree_centrality(G)[node] for node in G.nodes()]

nx.draw(G, pos, with_labels = True, node_size = node_sizes, font_size = 8, font_weight = 'bold', alpha = 0.7, arrowsize = 8,
       edge_color = 'gray', arrows = True, arrowstyle = '-|>')

edge_labels = {(u, v): f"{d['weight']}s" for u, v, d in G.edges(data = True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels = edge_labels, font_size = 6, label_pos = 0.5)


plt.title("Network Graph of Rapid Rail Stations")
plt.show()

num_vertices, num_edges
```

### 3. Calculate the Degree of the Network ###

```python
# Calculate the degree for each station
# Degree is the number of connections each station has (inbound + outbound)
from_station_counts = rail_network_df['from_station'].value_counts()
to_station_counts = rail_network_df['to_station'].value_counts()

# Sum inbound and outbound connections for each station
total_degree = from_station_counts.add(to_station_counts, fill_value=0).sort_values(ascending=False)

# Convert to DataFrame for better readability
degree_df = total_degree.reset_index()
degree_df.columns = ['station', 'degree']

# Display to the user
degree_df.head(10)
```

```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a graph using NetworkX
G = nx.Graph()

# Add edges to the graph from the dataset
for _, row in rail_network_df.iterrows():
    G.add_edge(row['from_station'], row['to_station'], weight=row['weight'])

# Calculate degree for each node
node_degrees = dict(G.degree())

# Map degree values to colors (higher degree, darker color)
node_colors = [node_degrees[node] for node in G.nodes()]

# Draw the network graph
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(G, seed=42)  # Spring layout for better visualization

# Draw nodes with color mapping based on degree
nx.draw_networkx_nodes(G, pos, node_size=100, node_color=node_colors, cmap=plt.cm.Blues)
nx.draw_networkx_edges(G, pos, alpha=0.3)

# Add labels for nodes (stations)
nx.draw_networkx_labels(G, pos, font_size=8)

# Add a color bar for degree levels
sm = plt.cm.ScalarMappable(cmap=plt.cm.Blues, norm=plt.Normalize(vmin=min(node_colors), vmax=max(node_colors)))
plt.colorbar(sm, label="Degree of Station")

# Title and show plot
plt.title("Rail Network Visualization by Degree")
plt.show()
```

In the rail network graph, each node represents a station, and edges represent connections between stations. Stations with higher degrees, indicated by darker colors, are central hubs with more connections to other stations. For example, TITIWANGSA, CHAN SOW LIN, and TUN RAZAK EXCHANGE have the highest degrees, highlighting them as significant transit points in the network. These key stations likely facilitate higher passenger flow or operational significance due to their numerous direct connections, making them essential for network connectivity and efficiency. Stations with lighter colors have fewer connections, representing smaller or less central stops in the network structure.

### 4. Determining the Minimum Spanning Tree (MST) ###

```python
# Compute the Minimum Spanning Tree (MST) of the network using Kruskal's algorithm
# Assuming the 'weight' column represents the cost or distance for the edges

mst = nx.minimum_spanning_tree(G, weight='weight')

# Extract the edges of the MST and their weights for easy viewing
mst_edges = [(u, v, d['weight']) for u, v, d in mst.edges(data=True)]
mst_df = pd.DataFrame(mst_edges, columns=['From Station', 'To Station', 'Weight'])
mst_df = mst_df.sort_values(by = 'Weight', ascending = False)

# Display the MST edges DataFrame
mst_df.head(10)
```

```python
# Visualizing the MST using a network graph plot
plt.figure(figsize=(15, 10))

# Creating a graph from the MST edges
MST_graph = nx.Graph()
MST_graph.add_weighted_edges_from(mst_edges)

# Drawing the graph with node labels and edge weights
pos = nx.spring_layout(MST_graph)
nx.draw(MST_graph, pos, with_labels=True, node_size=500, node_color='lightblue', font_size=10, font_weight='bold', edge_color='gray')
nx.draw_networkx_edge_labels(MST_graph, pos, edge_labels={(u, v): d['weight'] for u, v, d in MST_graph.edges(data=True)})

plt.title("Minimum Spanning Tree (MST) of the Network")
plt.show()
```

### The importance of MST to this Network ###

The Minimum Spanning Tree (MST) is crucial in this rail network as it identifies the most efficient subset of connections that link all stations with the minimum total cost or distance, ensuring full connectivity. By highlighting the paths that minimize infrastructure development or operational costs, the MST helps optimize resource allocation, reduce redundancy, and ensure cost-effective maintenance. This is particularly important for network planning, where minimizing construction expenses and maximizing connectivity can lead to a more sustainable and economical transportation system. The MST serves as an essential tool for decision-makers to balance connectivity with budget constraints.

### 5. Determine the shortest path of the rail network ###

Calculate the shortest path for each pair of nodes in the network using Dijkstra's algorithm assuming 'weight' represents the distance or cost for the edges

```python
# Compute shortest paths between all pairs of stations
shortest_paths = dict(nx.all_pairs_dijkstra_path_length(G, weight='weight'))

# Convert to a DataFrame for easy viewing - focusing on shortest path distances between all pairs
shortest_path_df = pd.DataFrame(shortest_paths).fillna(float('inf'))  # Use 'inf' where no path exists

# Display the shortest path distance matrix
shortest_path_df.head(10)
```

Here is a subset of the shortest path distance matrix for this network, where each entry represents the minimum travel weight (distance or cost) between two stations. For example, the shortest path between AMPANG and CAHAYA has a weight of 300, while the shortest path between AMPANG and HANG TUAH is 2700.

### Visualize the shortest path between two stations ###

Calculate the shortest path from AMPANG to PUDU using Dijkstra's algorithm

```python
shortest_path_ampang_pudu = nx.shortest_path(G, source='AMPANG', target='PUDU', weight='weight')
shortest_path_length = nx.shortest_path_length(G, source='AMPANG', target='PUDU', weight='weight')

# Visualize the shortest path from AMPANG to PUDU
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, seed=42)

# Draw the full network in the background for context
nx.draw_networkx_nodes(G, pos, node_size=100, node_color='lightgray', alpha=0.6)
nx.draw_networkx_edges(G, pos, edge_color='gray', alpha=0.3)
nx.draw_networkx_labels(G, pos, font_size=8, font_color='black')

# Highlight the shortest path
path_edges = list(zip(shortest_path_ampang_pudu, shortest_path_ampang_pudu[1:]))
nx.draw_networkx_nodes(G, pos, nodelist=shortest_path_ampang_pudu, node_size=150, node_color='lightblue')
nx.draw_networkx_edges(G, pos, edgelist=path_edges, edge_color='blue', width=2)
nx.draw_networkx_labels(G, pos, font_size=8)

# Add the path length as a title
plt.title(f"Shortest Path from AMPANG to PUDU (Distance: {shortest_path_length})")
plt.show()
```

This graph displays the shortest path from AMPANG to PUDU within the complete network of stations. The full network is shown in light gray, providing context for the relative positions of all stations and connections, while the blue nodes and edges represent the most efficient route between AMPANG and PUDU, with a total distance of 2400. This highlighted path demonstrates how specific connections minimize travel cost or time, making it the optimal route for passengers aiming for efficiency. This insight is valuable for both passengers seeking the quickest travel option and for network operators who may prioritize maintenance and service along such frequently used paths. By identifying and understanding shortest paths, planners can enhance both user experience and operational reliability in key areas of the network.

### 6. Determine the Clustering Coefficient of the rail network ###

The clustering coefficient measures the degree to which vertices in the graph tend to cluster together.

```python
# Calculate the clustering coefficient for each node in the network
clustering_coefficients = nx.clustering(G)

# Convert the clustering coefficients into a DataFrame for easier viewing
clustering_df = pd.DataFrame(list(clustering_coefficients.items()), columns=['Station', 'Clustering Coefficient'])

# Display the DataFrame
clustering_df.head(10)
```

The clustering coefficients for the stations in this network, as shown in the first ten entries, are all 0. This indicates that there is no triangle formation among neighboring nodes in these portions of the network, meaning that any two neighbors of a station are not directly connected to each other.

In general, a clustering coefficient of 0 for a node suggests a tree-like or linear structure, with little interconnectedness among neighboring nodes. This pattern is common in transportation networks where stations primarily connect linearly rather than in closed loops.

### 7. Determine the Degree Centrality of the rail network ###

```python
# Calculate degree centrality
degree_centrality = nx.degree_centrality(G)

# Create a DataFrame to display degree centrality values for all stations
degree_df = pd.DataFrame(degree_centrality.items(), columns=['Station', 'Degree Centrality'])

# Extract the top 10 stations with the highest degree centrality for display
top_10_degree = degree_df.sort_values(by='Degree Centrality', ascending=False).head(10)
top_10_degree
```

```python
# Normalize degree centrality values for coloring nodes
degree_values = list(degree_centrality.values())
min_degree = min(degree_values)
max_degree = max(degree_values)

# Normalize centrality for color mapping
node_colors_degree = [(value - min_degree) / (max_degree - min_degree) for value in degree_values]

# Draw the graph with nodes colored according to degree centrality
plt.figure(figsize=(14, 14))
pos = nx.spring_layout(G, seed=42)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=300,
    font_size=8,
    font_weight='bold',
    alpha=0.7,
    arrowsize=8,
    edge_color='gray',
    arrows=True,
    node_color=node_colors_degree,
    cmap=plt.cm.inferno
)

plt.title("Network Graph of Rapid Rail Stations with Degree Centrality")
plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.inferno), label='Degree Centrality')
plt.show()
```

In the graph colored by degree centrality, nodes representing key stations such as MASJID JAMEK, CHAN SOW LIN, and TUN RAZAK EXCHANGE would be colored with more intense shades. This indicates their high degree centrality values, showing that these stations have many direct connections and serve as major hubs within the rail network. The inferno color map, ranging from dark to bright shades, was used to emphasize the relative connectivity of each station — stations with more connections appear more vibrant, while those with fewer connections are shaded lighter.

### 8. Determine the Closeness Centrality of the rail network ###

Closeness centrality measures how quickly a vertex can reach all other vertices in the network.

```python
# Calculate closeness centrality for each node in the network
closeness_centrality = nx.closeness_centrality(G)

# Convert to a DataFrame and sort from highest to lowest
closeness_centrality_df = pd.DataFrame(closeness_centrality.items(), columns=['Station', 'Closeness Centrality'])
closeness_centrality_df = closeness_centrality_df.sort_values(by='Closeness Centrality', ascending=False).reset_index(drop=True)

# Display the sorted DataFrame
closeness_centrality_df.head(10)
```

Closeness centrality measures how close a station is to all other stations in the network, with higher values indicating that a station can reach others with relatively fewer steps or lower travel cost. Stations like TUN RAZAK EXCHANGE and BUKIT BINTANG have high centrality, suggesting they are more centrally located within the network, making them critical for efficient travel and connectivity. 

```python
# Normalize closeness centrality values for coloring nodes
closeness_values = list(closeness_centrality.values())
min_closeness = min(closeness_values)
max_closeness = max(closeness_values)

# Normalize centrality for color mapping
node_colors_closeness = [(value - min_closeness) / (max_closeness - min_closeness) for value in closeness_values]

# Draw the graph with nodes colored according to closeness centrality
plt.figure(figsize=(14, 14))
pos = nx.spring_layout(G, seed=42)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=300,
    font_size=8,
    font_weight='bold',
    alpha=0.7,
    arrowsize=8,
    edge_color='gray',
    arrows=True,
    node_color=node_colors_closeness,
    cmap=plt.cm.cividis
)

plt.title("Network Graph of Rapid Rail Stations with Closeness Centrality")
plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.cividis), label='Closeness Centrality')
plt.show()
```

In the closeness centrality visualized network graph, key stations such as MASJID JAMEK, CHAN SOW LIN, HANG TUAH, and TUN RAZAK EXCHANGE are highlighted with the lighter colors, indicating their high closeness centrality values. These stations are well-positioned within the network, providing efficient access to other stations, which makes them crucial for minimizing overall travel distances. Their central placement within the rail system suggests that passengers can reach other destinations more directly and quickly from these locations. In contrast, stations at the periphery, with deeper colors, have lower closeness centrality, implying longer average paths to other parts of the network. These differences help identify which stations are most efficient for travelers aiming to move across the network with minimal transit times.

### 9. Determine the Betweeness Centrality of the rail network ###

```python
# Calculate betweenness centrality for each node in the network
betweenness_centrality = nx.betweenness_centrality(G, weight='weight')

# Convert to a DataFrame and sort from highest to lowest
betweenness_centrality_df = pd.DataFrame(betweenness_centrality.items(), columns=['Station', 'Betweenness Centrality'])
betweenness_centrality_df = betweenness_centrality_df.sort_values(by='Betweenness Centrality', ascending=False).reset_index(drop=True)

# Display the sorted DataFrame
betweenness_centrality_df.head(10)
```

Betweenness centrality measures the extent to which a station acts as an intermediary along the shortest paths between other stations. High betweenness values, as seen in CHAN SOW LIN and MASJID JAMEK, indicate that these stations are crucial transit points in the network, frequently traversed in optimal routing paths. These stations are key for network flow and resilience, as disruptions here could significantly impact overall connectivity.

```python
# Normalize betweenness centrality values for coloring nodes
betweenness_values = list(betweenness_centrality.values())
min_betweenness = min(betweenness_values)
max_betweenness = max(betweenness_values)

# Normalize centrality for color mapping
node_colors_betweenness = [(value - min_betweenness) / (max_betweenness - min_betweenness) for value in betweenness_values]

# Draw the graph with nodes colored according to betweenness centrality
plt.figure(figsize=(14, 14))
pos = nx.spring_layout(G, seed=42)

nx.draw(
    G,
    pos,
    with_labels=True,
    node_size=300,
    font_size=8,
    font_weight='bold',
    alpha=0.7,
    arrowsize=8,
    edge_color='gray',
    arrows=True,
    node_color=node_colors_betweenness,
    cmap=plt.cm.plasma
)

plt.title("Network Graph of Rapid Rail Stations with Betweenness Centrality")
plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.plasma), label='Betweenness Centrality')
plt.show()
```

In the betweenness centrality visualized network graph, key stations like CHAN SOW LIN, HANG TUAH, TUN RAZAK EXCHANGE, and MASJID JAMEK are highlighted with the most lighter colors, indicating their high betweenness centrality values. These stations serve as crucial transfer points in the network, acting as intermediaries for passenger flow between multiple major routes. As a result, they play a vital role in the overall connectivity and efficiency of the rail system. Stations with lower betweenness, indicated by deeper colors, are typically endpoints or have fewer direct connections, suggesting they are less critical for inter-route transfers compared to the highly central hubs. This analysis helps identify which stations are pivotal in ensuring smooth travel across the entire network.

### 10. Determine the Eigenvector Centrality of the rail network ###

```python
# Create directed graph
G = nx.DiGraph()

# Add edges from the DataFrame
for _, row in rail_network_df.iterrows():
    G.add_edge(row['from_station'], row['to_station'], weight=row['weight'])

# Calculate eigenvector centrality
eigenvector_centrality = nx.eigenvector_centrality(G, max_iter=1000)
# Convert to a DataFrame and sort from highest to lowest
eigenvector_centrality_df = pd.DataFrame(eigenvector_centrality.items(), columns=['Station', 'Eigenvector Centrality'])
eigenvector_centrality_df = eigenvector_centrality_df.sort_values(by='Eigenvector Centrality', ascending=False).reset_index(drop=True)

eigenvector_centrality_df.head(10)
```

```python
# Normalize eigenvector centrality values for coloring nodes
centrality_values = list(eigenvector_centrality.values())
min_centrality = min(centrality_values)
max_centrality = max(centrality_values)

# Normalize centrality for color mapping
node_colors = [(value - min_centrality) / (max_centrality - min_centrality) for value in centrality_values]

# Draw the graph with nodes colored according to eigenvector centrality
plt.figure(figsize=(14, 14))
pos = nx.spring_layout(G, seed=42)

nx.draw(
    G, 
    pos, 
    with_labels=True, 
    node_size=300, 
    font_size=8, 
    font_weight='bold', 
    alpha=0.7, 
    arrowsize=8, 
    edge_color='gray', 
    arrows=True, 
    node_color=node_colors, 
    cmap=plt.cm.viridis
)

plt.title("Network Graph of Rapid Rail Stations with Eigenvector Centrality Coloring")
plt.colorbar(plt.cm.ScalarMappable(cmap=plt.cm.viridis), label='Eigenvector Centrality')
plt.show()
```

In the visualized network graph, the nodes represent different rail stations, with their colors indicating their eigenvector centrality — a measure of influence within the network. Key stations such as CHAN SOW LIN, TUN RAZAK EXCHANGE, HANG TUAH, and BUKIT BINTANG are among the most central and influential, as highlighted by their lighter colors. These stations act as critical hubs, facilitating connections across multiple routes, which increases their importance within the network. In contrast, stations with deeper colors, such as PUTRAJAYA SENTRAL and SUNWAY LAGOON, exhibit lower centrality, indicating that they have fewer direct or indirect connections compared to the more central nodes. This difference suggests that the highly connected nodes play a pivotal role in maintaining the overall connectivity and efficiency of the rapid rail system.
