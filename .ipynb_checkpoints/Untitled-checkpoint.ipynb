```python
import requests
import pandas as pd
import zipfile
import io
import os

# Define the API endpoint for Rapid Rail Explorer data
api_url = "https://api.data.gov.my/gtfs-static/prasarana?category=rapid-rail-kl"

# Make the request to the API
response = requests.get(api_url)

# Check if the request was successful
if response.status_code == 200:
    # Load the content into a BytesIO object
    z = zipfile.ZipFile(io.BytesIO(response.content))
    # Extract all files into the current directory
    z.extractall()
    print("Files extracted successfully.")
    # Load specific GTFS files into pandas DataFrames
    stops_df = pd.read_csv('stops.txt')
    trips_df = pd.read_csv('trips.txt')
    stop_times_df = pd.read_csv('stop_times.txt')
    routes_df = pd.read_csv('routes.txt')
    # Save each DataFrame to a CSV file
    stops_df.to_csv("stops.csv", index=False)
    trips_df.to_csv("trips.csv", index=False)
    stop_times_df.to_csv("stop_times.csv", index=False)
    routes_df.to_csv("routes.csv", index=False)
    print("Data successfully saved to CSV files.")
else:
    print(f"Failed to retrieve data. Status code: {response.status_code}")

# Notes:
# - This script retrieves data from the GTFS Static API for Rapid Rail.
# - The extracted files are saved in CSV format for easier analysis.
```

```python
import pandas as pd

# Load the extracted CSV files
stops_df = pd.read_csv('stops.csv')
trips_df = pd.read_csv('trips.csv')
stop_times_df = pd.read_csv('stop_times.csv')
routes_df = pd.read_csv('routes.csv')
```

```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Create an empty directed graph
G = nx.DiGraph()

# Add nodes for each stop
for index, row in stops_df.iterrows():
    G.add_node(row['stop_id'], name=row['stop_name'])

# Add edges using stop_times information
# Edges represent connections between consecutive stops in a trip
for trip_id in stop_times_df['trip_id'].unique():
    trip_stops = stop_times_df[stop_times_df['trip_id'] == trip_id]
    trip_stops = trip_stops.sort_values(by='stop_sequence')
    
    # Iterate through consecutive stops in the trip
    for i in range(len(trip_stops) - 1):
        from_stop = trip_stops.iloc[i]['stop_id']
        to_stop = trip_stops.iloc[i + 1]['stop_id']
        
        # Add edge between consecutive stops
        if G.has_edge(from_stop, to_stop):
            # If edge already exists, you can increment weight or skip it
            G[from_stop][to_stop]['weight'] += 1
        else:
            G.add_edge(from_stop, to_stop, weight=1)

print(f"Number of nodes: {G.number_of_nodes()}")
print(f"Number of edges: {G.number_of_edges()}")
```

```python
plt.figure(figsize=(12, 12))

# Draw the graph
pos = nx.spring_layout(G, seed=42)  # Position nodes using a layout algorithm
nx.draw_networkx_nodes(G, pos, node_size=50, node_color='blue')
nx.draw_networkx_edges(G, pos, arrowstyle='->', arrowsize=10, edge_color='gray')
nx.draw_networkx_labels(G, pos, font_size=8)

plt.title("Rapid Rail Network Graph")
plt.show()
```

```python
degree_centrality = nx.degree_centrality(G)
print(sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:10])
```

```python
source = 'valid_stop_id_1'  # Replace with an actual stop_id from stops_df
target = 'valid_stop_id_2'  # Replace with another stop_id from stops_df

try:
    shortest_path = nx.shortest_path(G, source=source, target=target, weight='weight')
    print("Shortest Path:", shortest_path)
except nx.NodeNotFound as e:
    print(e)
```

```python
all_stops_in_graph = set(G.nodes)
stop_ids_in_data = set(stop_times_df['stop_id'].unique())

missing_stops = stop_ids_in_data - all_stops_in_graph
if missing_stops:
    print(f"Warning: Some stops are not found in the graph: {missing_stops}")
else:
    print("All stops are accounted for in the graph.")
```

```python
import pandas as pd

# Load the extracted CSV files
stops_df = pd.read_csv('stops.csv')
trips_df = pd.read_csv('trips.csv')
stop_times_df = pd.read_csv('stop_times.csv')
routes_df = pd.read_csv('routes.csv')

# Merge stop_times with stops based on stop_id
stops_combined = pd.merge(stop_times_df, stops_df, on='stop_id', how='left')

# Merge the result with trips based on trip_id
trips_combined = pd.merge(stops_combined, trips_df, on='trip_id', how='left')

# Merge the result with routes based on route_id
final_combined_df = pd.merge(trips_combined, routes_df, on='route_id', how='left')

# Save the final combined DataFrame to a CSV file
final_combined_df.to_csv('combined_rapid_rail_data.csv', index=False)

print("All CSV files successfully combined into 'combined_rapid_rail_data.csv'")
```
