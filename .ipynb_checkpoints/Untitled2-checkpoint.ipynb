```python
import pandas as pd

# Step 1: Load the dataset
file_path = 'combined_rapid_rail_data.csv'
data = pd.read_csv(file_path)

# Step 2: Select relevant columns for network analysis
columns_to_keep = ['trip_id', 'stop_id', 'stop_name', 'stop_sequence']
network_data = data[columns_to_keep]

# Step 3: Sort data by trip_id and stop_sequence to ensure proper order for edge construction
network_data = network_data.sort_values(by=['trip_id', 'stop_sequence'])

# Step 4: Construct edges by pairing consecutive stations within the same trip
edges = network_data.groupby('trip_id').apply(
    lambda x: pd.DataFrame({
        'from_stop_id': x['stop_id'].iloc[:-1].values,
        'to_stop_id': x['stop_id'].iloc[1:].values,
        'from_stop_name': x['stop_name'].iloc[:-1].values,
        'to_stop_name': x['stop_name'].iloc[1:].values
    })
).reset_index(drop=True)

# The edges dataframe is now ready for network analysis
```

```python
# Save the cleaned dataset as a CSV file
output_file_path = 'C:/Users/kamar/Network_Science/network_analysis_data.csv'
edges.to_csv(output_file_path, index=False)

output_file_path
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Create a directed graph from the edges
G = nx.DiGraph()
for _, row in edges.iterrows():
    G.add_edge(row['from_stop_name'], row['to_stop_name'])

# Plot the network graph
plt.figure(figsize=(12, 8))
pos = nx.spring_layout(G, k=0.3)  # Positioning the nodes
nx.draw(G, pos, with_labels=True, node_size=500, font_size=8, alpha=0.7, arrowsize=10)
plt.title("Network Graph of Rapid Rail Stations")
plt.show()
```

```python
# Performing some basic network analysis on the graph

# Calculate and display basic graph statistics
num_nodes = G.number_of_nodes()
num_edges = G.number_of_edges()
density = nx.density(G)

# Calculate the degree centrality of the graph
degree_centrality = nx.degree_centrality(G)

# Calculate the betweenness centrality
betweenness_centrality = nx.betweenness_centrality(G)

# Find the shortest path length between all pairs
average_shortest_path_length = None
if nx.is_connected(G.to_undirected()):
    average_shortest_path_length = nx.average_shortest_path_length(G.to_undirected())

# Minimum Spanning Tree (for undirected graph version)
MST = nx.minimum_spanning_tree(G.to_undirected())

# Calculate the shortest path between all pairs (only if the graph is connected)
shortest_paths = dict(nx.all_pairs_shortest_path_length(G.to_undirected()))

# Calculate the clustering coefficient for the graph
clustering_coefficient = nx.average_clustering(G.to_undirected())

# Calculate closeness centrality
closeness_centrality = nx.closeness_centrality(G)
    
# Collect results into a dictionary
network_analysis_results = {
    "Number of Nodes": num_nodes,
    "Number of Edges": num_edges,
    "Graph Density": density,
    "Average Shortest Path Length": average_shortest_path_length,
    "Top 5 Nodes by Degree Centrality": sorted(degree_centrality.items(), key=lambda x: x[1], reverse=True)[:5],
    "Top 5 Nodes by Betweenness Centrality": sorted(betweenness_centrality.items(), key=lambda x: x[1], reverse=True)[:5],
    "Top 5 Minimum Spanning Tree Edges": list(MST.edges)[:5],
    "Average Clustering Coefficient": clustering_coefficient,
    "Top 5 Nodes by Closeness Centrality": sorted(closeness_centrality.items(), key=lambda x: x[1], reverse=True)[:5],
    "Sample Shortest Paths": {node: paths for node, paths in list(shortest_paths.items())[:1]}  # Display sample paths
}

network_analysis_results
```
