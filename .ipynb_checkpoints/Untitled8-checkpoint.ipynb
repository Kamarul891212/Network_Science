```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt

file_path = "facebook_combined.txt"
data = pd.read_csv(file_path, sep = " ", header = None, names = ["source", "target"])

data["weight"] = 1

data_aggregated = data.groupby(["source", "target"]).sum().reset_index()

G = nx.Graph()

for _, row in data_aggregated.iterrows():
    G.add_edge(row["source"], row["target"], weight = row["weight"])
    
plt.figure(figsize = (12, 12))
pos = nx.spring_layout(G, seed = 42)
weights = nx.get_edge_attributes(G, 'weight')

nx.draw(G, pos, node_size = 20, edge_color = "gray", alpha = 0.6, with_labels = False)

nx.draw_networkx_edges(G, pos, edge_color = "blue", alpha = 0.4, width = [w for w in weights.values()],)

plt.title("Facebook Network Graph", fontsize = 16)
plt.show()
```

```python
mst = nx.minimum_spanning_tree(G, weight = 'weight')

plt.figure(figsize = (12, 12))
pos = nx.spring_layout(mst, seed = 42)
weights_mst = nx.get_edge_attributes(mst, 'weight')

nx.draw(mst, pos, node_size = 20, edge_color = "gray", alpha = 0.6, with_labels = False)

nx.draw_networkx_edges(mst, pos, edge_color = "red", alpha = 0.7, width = [w for w in weights_mst.values()])

plt.title("Minimum Spanning Tree of the Facebook Network", fontsize = 16)
plt.show()
```

```python
degree_centrality = nx.degree_centrality(G)

degree_centrality_df = pd.DataFrame({
    'node': degree_centrality.keys(),
    'degree_centrality': degree_centrality.values()
})

plt.figure(figsize = (10, 6))
plt.hist(degree_centrality_df['degree_centrality'], bins = 30, alpha = 0.7, edgecolor = 'black')
plt.title('Degree Centrality Distrbution', fontsize = 16)
plt.xlabel('Degree Centrality', fontsize = 12)
plt.ylabel('Frequency', fontsize = 12)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()

degree_centrality_df[:10]
```

```python
closeness_centrality = nx.closeness_centrality(G)

closeness_centrality_df = pd.DataFrame({
    'node': closeness_centrality.keys(),
    'closeness_centrality': closeness_centrality.values()
})

plt.figure(figsize = (10, 6))
plt.hist(closeness_centrality_df['closeness_centrality'], bins = 30, alpha = 0.7, edgecolor = 'black')
plt.title('Closeness Centrality Distrbution', fontsize = 16)
plt.xlabel('Closeness Centrality', fontsize = 12)
plt.ylabel('Frequency', fontsize = 12)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()

closeness_centrality_df[:10]
```

```python
betweenness_centrality = nx.betweenness_centrality(G)

betweenness_centrality_df = pd.DataFrame({
    'node': betweenness_centrality.keys(),
    'betweenness_centrality': betweenness_centrality.values()
})

plt.figure(figsize = (10, 6))
plt.hist(betweenness_centrality_df['betweenness_centrality'], bins = 30, alpha = 0.7, edgecolor = 'black')
plt.title('Betweenness Centrality Distrbution', fontsize = 16)
plt.xlabel('Betweenness Centrality', fontsize = 12)
plt.ylabel('Frequency', fontsize = 12)
plt.grid(axis = 'y', linestyle = '--', alpha = 0.7)
plt.show()

betweenness_centrality_df[:10]
```
