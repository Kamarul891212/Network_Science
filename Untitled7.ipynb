```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# load data from Facebook dataset file
file_path = "facebook_combined.txt"

# create empty graph
G = nx.Graph()

# read the edges from dataset
with open(file_path, 'r') as f:
    for line in f:
        edge = tuple(map(int, line.strip().split()))
        G.add_edge(*edge)
```

```python
# print basic information
print("Number of nodes:", G.number_of_nodes())
print("Number of edges:", G.number_of_edges())

# Display degree distribution
degrees = [degree for _, degree in G.degree()]
plt.hist(degrees, bins = 50)
plt.title("Degree Distribution")
plt.xlabel("Degree")
plt.ylabel("Frequency")
plt.show()
```

```python
# Find a node based on highest degree and centrality

# highest degree
highest_degree_node = max(dict(G.degree()).items(), key = lambda x: x[1])[0]
ego_node_degree = highest_degree_node

# betweenness centrality
betweenness = nx.betweenness_centrality(G)
ego_node_betweenness = max(betweenness, key = betweenness.get)

# closeness centrality
closeness = nx.closeness_centrality(G)
ego_node_closeness = max(closeness, key = closeness.get)

# eigenvector centrality
eigenvector = nx.eigenvector_centrality(G)
ego_node_eigenvector = max(eigenvector, key = eigenvector.get)
```

```python
print("Highest degree node:", ego_node_degree)
print("Highest betweenness centrality node:", ego_node_betweenness)
print("Highest closeness centrality node:", ego_node_closeness)
print("Highest eigenvector centrality node:", ego_node_eigenvector)
```

```python
# Select Nodes, Edges and Weight

# extract a subgraph of a specific node's neighbour (ego network)
ego_node = 1912
ego_graph = nx.ego_graph(G, ego_node)

# add weight (based on degree)
for u, v in ego_graph.edges():
    ego_graph[u][v]['weight'] = ego_graph.degree[u] + ego_graph.degree[v]
```

```python
# visualize the graph
pos = nx.spring_layout(ego_graph)
weights = nx.get_edge_attributes(ego_graph, 'weight')

plt.figure(figsize = (8, 8))
nx.draw(ego_graph, pos, with_labels = True, node_size = 500,
       font_size = 10)
nx.draw_networkx_edge_labels(ego_graph, pos, edge_labels = weights)
plt.title(f"Ego Network of Node {ego_node}")
plt.show()
```
